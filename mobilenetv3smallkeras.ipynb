```python
from PIL import Image
image = Image.open("./lighter.jpg")
image = image.resize((320,320),Image.ANTIALIAS)
image.save(fp="newimage.png")
```

```python
import numpy as np
from imageio import imread
from PIL import Image

data = np.empty((1, 320, 320, 3))
data[0] = imread('newimage.png')
```

```python
from keras.applications.mobilenet_v2 import preprocess_input

data = preprocess_input(data)
```

```python
predictions = model.predict(data)
print('Shape: {}'.format(predictions.shape))
```

```python
from keras.applications.mobilenet_v2 import decode_predictions

for name, desc, score in decode_predictions(predictions)[0]:
    print('- {} ({:.2f}%%)'.format(desc, 100 * score))
```
