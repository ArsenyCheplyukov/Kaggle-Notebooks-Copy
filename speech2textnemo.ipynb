```python
BATCH_SIZE = 4
```

```python
import nemo.collections.asr as nemo_asr
model = nemo_asr.models.ASRModel.from_pretrained(model_name="QuartzNet15x5Base-En")
```

```python
nemo_asr.models.ASRModel.list_available_models()
```

```python
from time import time
start = time()
print(model.transcribe(paths2audio_files=[
    "../input/vidtimit-audiovideo-dataset/fadg0/fadg0/audio/si649.wav",
    "../input/vidtimit-audiovideo-dataset/fadg0/fadg0/audio/sx109.wav",
    "../input/vidtimit-audiovideo-dataset/fadg0/fadg0/audio/sx19.wav",
    "../input/vidtimit-audiovideo-dataset/fadg0/fadg0/audio/sx199.wav"], batch_size=BATCH_SIZE, logprobs=False))
print(f"THE FINAL TIME IS: {time() - start}")
```
